## Validacion de campos y columnas

```python
import pandas as pd

def validar_dataframe(df: pd.DataFrame, required_columns: list, nullable_columns: set):
    """
    Valida que el DataFrame cumpla con:
    1. Todas las columnas requeridas existen.
    2. Solo las columnas permitidas contienen valores nulos.
    """
    try:
        # 1. Verificar columnas faltantes
        missing = set(required_columns) - set(df.columns)
        if missing:
            raise ValueError(f"Faltan columnas requeridas: {missing}")

        # 2. Verificar que no existan nulos en columnas no permitidas
        cols_no_nullable = [c for c in required_columns if c not in nullable_columns]
        nulls_invalidos = df[cols_no_nullable].isnull().any()
        if nulls_invalidos.any():
            raise ValueError(f"Se encontraron valores nulos en columnas no permitidas: "
                             f"{list(nulls_invalidos[nulls_invalidos].index)}")

    except Exception as e:
        print("❌ Validación fallida:", e)
        raise  # detiene el proceso
    else:
        print("✅ DataFrame válido: todas las columnas presentes y nulos correctos.")


# =======================
# Ejemplos de uso
# =======================

# --- df1 ---
required_columns_df1 = [
    "id", "Fecha.Expedicion", "Cancelacion", "Gestionable", "Retencion", "TIPO", 
    "ANO_MES", "Target", "Fecha.Proceso", "Disponible.Avances", "Limite.Avances",
    "Total.Intereses", "Saldos.Mes.Ant", "Pagos.Mes.Ant", "Vtas.Mes.Ant",
    "Edad.Mora", "Limite.Cupo", "Pago.del.Mes", "Pago.Minimo", 
    "Vr.Mora", "Vr.Cuota.Manejo", "Saldo"
]
nullable_columns_df1 = {"Cancelacion", "Gestionable", "Retencion", "TIPO", "ANO_MES"}
validar_dataframe(df1, required_columns_df1, nullable_columns_df1)


# --- df2 ---
required_columns_df2 = [
    'id', 'categoria', 'segmento', 'edad', 'nivel_educativo',
    'estado_civil', 'Genero', 'PAC', 'contrato', 'estrato'
]
nullable_columns_df2 = {"estrato"}
validar_dataframe(df2, required_columns_df2, nullable_columns_df2)


# --- df3 ---
required_columns_df3 = ['id', 'cuota_monetaria', 'sub_vivenda', 'bono_lonchera']
nullable_columns_df3 = {"bono_lonchera"}
validar_dataframe(df3, required_columns_df3, nullable_columns_df3)
```

## Validacion de fechas y campos minimos por documento

```python
class DataQualityError(Exception):
    pass

def validar_filas_minimas(dfs: dict, min_rows: dict):
    """
    dfs: {"df1": df1, "df2": df2, "df3": df3}
    min_rows: {"df1": 50001, "df2": 50001, "df3": 50001}
    """
    for name, df in dfs.items():
        print(f"{name}.shape = {df.shape}")
        req = min_rows.get(name, 0)
        if df.shape[0] < req:
            raise DataQualityError(f"{name} tiene {df.shape[0]} filas < mínimo requerido {req}")
    print("✅ Filas mínimas: OK.")

def validar_regla_fechas_df1(df1: pd.DataFrame,
                             col_expedicion="Fecha.Expedicion",
                             col_proceso="Fecha.Proceso"):
    """
    Convierte las columnas de fecha con formato dd/mm/YYYY
    y valida que Fecha.Proceso >= Fecha.Expedicion en df1.
    """
    if col_expedicion not in df1.columns or col_proceso not in df1.columns:
        raise DataQualityError(f"Faltan columnas '{col_expedicion}' y/o '{col_proceso}' en df1")

    df1[col_proceso] = pd.to_datetime(df1[col_proceso], format="%d/%m/%Y", errors="coerce")
    df1[col_expedicion] = pd.to_datetime(df1[col_expedicion], format="%d/%m/%Y", errors="coerce")

    # Validar NaT
    if df1[col_proceso].isna().any() or df1[col_expedicion].isna().any():
        raise DataQualityError("❌ Existen fechas inválidas tras la conversión con formato dd/mm/YYYY")

    # Validar regla
    bad = ~(df1[col_proceso] >= df1[col_expedicion])
    if bad.any():
        n_bad = int(bad.sum())
        muestra_idx = df1.index[bad][:5].tolist()
        raise DataQualityError(
            f"'{col_proceso}' no es mayor que '{col_expedicion}' en {n_bad} filas. "
            f"Ejemplos de índices: {muestra_idx}"
        )

    print("✅ Regla de fechas (Fecha.Proceso >= Fecha.Expedicion): OK.")


# ======================
# Uso
# ======================
dfs = {"df1": df1, "df2": df2, "df3": df3}
min_rows = {"df1": 50001, "df2": 50001, "df3": 50001}

# 1) Validar mínimo de filas
validar_filas_minimas(dfs, min_rows)

# 2) Validar fechas en df1
validar_regla_fechas_df1(df1)
```
